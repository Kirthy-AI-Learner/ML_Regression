```python
import  pandas as pd
```

```python
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
```

```python
dataset
```

```python
indep = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep= dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test = train_test_split(indep,dep,random_state=0,test_size=1/3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['gini','entropy'],
              'splitter':['best','random'],
              'max_features':['sqrt','log2'],
             'class_weight':['balanced']}
DC=DecisionTreeClassifier()
grid = GridSearchCV(DC,param_grid,verbose=3,cv=5,n_jobs=-1,refit=True,scoring='f1_weighted')
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_test,y_pred,average='weighted')
print('best params and f1 score',grid.best_params_,score)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
