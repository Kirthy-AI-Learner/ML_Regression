```python
import pandas as pd
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independent_variable=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
dependent_variable=dataset[['charges']]
```

```python
independent_variable
```

```python
dependent_variable
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['absolute_error','squared_error','friedman_mse','poisson'],
           'splitter':['best','random'],
           'max_features':['sqrt','log2']}
grid = GridSearchCV(DecisionTreeRegressor(),param_grid,n_jobs=-1,verbose=3,refit=True)
grid.fit(independent_variable,dependent_variable)
re=grid.cv_results_
```
