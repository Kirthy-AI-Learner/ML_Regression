```python
import pandas as pd
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independent_variable=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
dependent_variable=dataset[['charges']]
```

```python
independent_variable
```

```python
dependent_variable
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV

param_grid ={'loss':['linear','exponential','square'],
            'n_estimators':[10,100],
             'random_state':[0,50]}
grid = GridSearchCV(AdaBoostRegressor(),param_grid,verbose=3,n_jobs=-1,refit=True)
grid.fit(independent_variable,dependent_variable)
re = grid.cv_results_
table = pd.DataFrame.from_dict(re)
```

```python
table
```

```python
print('Best R_Score value',format(grid.best_params_))
```

```python
grid_pred = grid.predict(independent_variable)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(dependent_variable,grid_pred)
```

```python
r_score
```
